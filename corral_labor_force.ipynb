```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import plotly.express as px
import plotly.io as pio
import plotly  
```

```python
pd_edu  = pd.read_csv("edudata.csv")
pd_edu_melt = pd_edu.melt(id_vars=['Country Name', 'Country Code', 'Indicator Code', 'Indicator Name'], var_name='Year', value_name='Value')
# data to csv




 
```

```python
pivoted_data = pd_edu_melt.pivot_table(index=['Country Name', 'Year'], columns='Indicator Name', values='Value')
pivoted_data.reset_index(inplace=True)
display(pivoted_data)
```

```python

#drop non int or float columns
pivoted_data = pivoted_data.drop(columns=['Country Name', 'Year'])
corralation_data = pivoted_data.corr()
corralation_data.dropna(axis=0, how='all', inplace=True)

#print each corralation line by line
for i in range(len(corralation_data)):
    for j in range(len(corralation_data)):
        if i != j:
            if corralation_data.iloc[i,j] < -0.4  or corralation_data.iloc[i,j] > 0.4:
                print(corralation_data.columns[i] + " and " + corralation_data.columns[j] + " have a correlation of " + str(corralation_data.iloc[i,j]))
```
