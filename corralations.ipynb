```python
import pandas as pd
```

```python
# Read the data from the CSV file
pd_edu = pd.read_csv('edudata.csv')
pd_life = pd.read_csv('life_expectancy.csv')
pd_gdp = pd.read_csv('GDP.csv')
pd_merged = pd.read_csv('merged_data.csv')
#display columna 
print(pd_merged.columns)
```

```python
# Ensure proper column name handling and filter only necessary columns for correlation
pd_merged.columns = pd_merged.columns.str.strip()

# List of unique indicator names
list_of_indicators = pd_merged['Indicator Name'].unique().tolist()

# Prepare to store results
correlation_results = []

# Iterate over each indicator name
for indicator in list_of_indicators:
    indicator_df = pd_merged[pd_merged['Indicator Name'] == indicator]
    
    # Remove columns with object dtype in indicator_df
    indicator_df = indicator_df.select_dtypes(exclude=['object'])
    
    # Calculate correlation
    correlation = indicator_df.corr()
    
    # Drop columns from correlation except 'Value'
    if 'Value' in correlation.columns:
        correlation = correlation[['Value']]
    
    correlation_results.append((indicator, correlation))

for i in correlation_results:
    print(i[0])
    display(i[1])

```
